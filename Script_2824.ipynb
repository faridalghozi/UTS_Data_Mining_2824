```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression
```

```python
df_car = pd.read_csv('toyota.csv')
df_car
```

```python
df_car.info()
```

```python
df_car.drop(columns="model")
```

```python
car2 = pd.get_dummies(df_car)
car2
```

```python
Y = df_car.price
X = car2.iloc[:,1:]
y = car2.iloc[:,2]
```

```python
X.shape
y.shape
```

```python
X_train, X_test, y_train, y_test = tts(X, y, train_size=0.7, test_size=0.3, random_state=10)
X_train
y
```

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
print(linreg.coef_)
print(linreg.intercept_)
```

```python
pred = linreg.predict(X_test)
pred
```

```python
y_test
```

```python
df_perbandingan = pd.DataFrame({'Actual':y_test, 'Predicted':pred})
df_perbandingan
```
